Q6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

Q9.  How can you read a file line by line and store its content in a list in Python?

```python
with open("example.txt", "w") as file:
    file.write("Hello, world!\n")
    file.write("This is a new line.\n")
    file.write("Another line.\n")
    file.write("Last line.\n")

lines = []
with open("example.txt", "r") as file:
    for line in file:
        lines.append(line.strip())

print(lines)
```

Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler
import os

# Clean up old log files for demo
for f in ["app.log", "app.log.1", "app.log.2", "app.log.3"]:
    if os.path.exists(f):
        os.remove(f)

# Custom handler that prints when a rollover happens
class PrintRotatingFileHandler(RotatingFileHandler):
    def doRollover(self):
        print(f"Rotating file: {self.baseFilename}")
        super().doRollover()

# Create logger
logger = logging.getLogger("JupyterLogger")
logger.setLevel(logging.INFO)

# Rotating handler with 1MB max size
handler = PrintRotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=3)
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)

# Log multiple messages (adjust number to exceed 1MB if needed)
for i in range(50000):  # Increase this number to trigger rotation in your environment
    logger.info(f"Log message number {i}")

# Check which log files exist
print("Current log files:", [f for f in os.listdir() if f.startswith("app.log")])

```

Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
import string

word_to_count = "line"
count = 0

with open("example.txt", "r") as file:
    for line in file:
        line_clean = line.translate(str.maketrans("", "", string.punctuation))
        words = line_clean.lower().split()
        count += words.count(word_to_count.lower())

print(f"The word '{word_to_count}' occurs {count} times in the file.")
```

Q23. Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging

logging.basicConfig(filename='file_errors.log',
                    level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

filename = "nonexistent_file.txt"

try:
    with open(filename, "r") as file:
        content = file.read()
        print(content)
except (FileNotFoundError, IOError) as e:
    print(f"Error: {e}")
    logging.error(f"Error occurred while handling file '{filename}'", exc_info=True)
```
