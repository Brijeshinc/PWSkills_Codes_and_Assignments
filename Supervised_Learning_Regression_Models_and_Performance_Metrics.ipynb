## **Theoretical Questions**

Q1. What is Simple Linear Regression (SLR)? Explain its purpose.

Ans. Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line. Its main purpose is to understand the relationship between variables and predict the value of Y based on X. It also helps in identifying trends and measuring how changes in X affect Y.

Q2. What are the key assumptions of Simple Linear Regression?

Ans. The key assumptions of Simple Linear Regression are:

Linearity – The relationship between the independent variable (X) and dependent variable (Y) is linear.

Independence – The observations (data points) are independent of each other.

Homoscedasticity – The variance of the errors (residuals) is constant for all values of X.

Normality – The residuals (errors) are normally distributed.

No significant outliers – The data should not contain extreme values that strongly influence the regression line.

Q3.  Write the mathematical equation for a simple linear regression model and explain each term.

Ans. Y = AX + B

     Y: Dependent variable

     X: Independent variable (predictor/input)

     B: Intercept (value of Y when X = 0)
     
     A: Slope (change in Y for a one-unit change in X)

Q5.  What is the method of least squares in linear regression?

Ans.The method of least squares in linear regression is used to find the best-fitting line by minimizing the errors between actual and predicted values. It works by minimizing the sum of the squares of the residuals (differences between observed and predicted values).

Q6.  What is Logistic Regression? How does it differ from Linear Regression?

Ans. Logistic Regression is a statistical method used for classification problems, where the output represents a probability between 0 and 1, commonly used to predict binary outcomes such as yes/no or pass/fail. It uses a sigmoid (S-shaped) function to model the relationship between variables. In contrast, Linear Regression is used to predict continuous numerical values and fits a straight line to the data. While Linear Regression outputs any real value, Logistic Regression restricts outputs to probabilities, making it suitable for classification tasks rather than prediction of exact quantities.

Q7.  Name and briefly describe three common evaluation metrics for regression models.

Ans. Three common evaluation metrics for regression models are Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE). MAE measures the average absolute difference between actual and predicted values, giving a simple idea of prediction error. MSE calculates the average of squared differences, penalizing larger errors more strongly. RMSE is the square root of MSE and provides the error in the same units as the target variable, making it easier to interpret.

Q8. What is the purpose of the R-squared metric in regression analysis?

Ans. The R-squared (R²) metric is used to measure how well a regression model fits the data. It represents the proportion of the variance in the dependent variable that is explained by the independent variable. An R² value ranges from 0 to 1, where a higher value indicates a better fit of the model. For example, an R² of 0.8 means that 80% of the variation in the output is explained by the model, while the remaining 20% is due to other factors or randomness.

Q9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

print("Slope (A):", model.coef_[0])
print("Intercept (B):", model.intercept_)
```
