Q1. What is a random variable in probability theory?

Ans. A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.

Q2. What are the types of random variables?

Ans. There are two main types of random variables:

Discrete random variables — take countable values (e.g., number of heads in coin tosses).

Continuous random variables — take values from a continuous range or interval (e.g., height, time, weight).

Q3. Explain the difference between discrete and continuous distributions.

Ans. A discrete distribution applies to random variables that take countable and distinct values, such as 0, 1, 2, or 3, and each value has an exact probability given by a probability mass function (PMF).

A continuous distribution applies to random variables that can take any value within a continuous range or interval, and probabilities are described using a probability density function (PDF).

Q4. What is a binomial distribution, and how is it used in probability?

Ans. A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials.

Q5. What is the standard normal distribution, and why is it important?

Ans. The standard normal distribution is a special case of the normal (bell-shaped) distribution with a mean of 0 and a standard deviation of 1. It is important because many natural and statistical phenomena follow a normal pattern, and converting any normal variable to the standard normal form (using z-scores) allows us to easily compute probabilities, compare different datasets, and use statistical methods such as hypothesis testing and confidence intervals. It essentially serves as a universal reference distribution in statistics.

Q6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans. The Central Limit Theorem (CLT) states that when independent, identically distributed random variables are averaged (or summed), their distribution approaches a normal distribution as the sample size becomes large, regardless of the original population’s shape.

Q7. What is the significance of confidence intervals in statistical analysis?

Ans. Confidence intervals provide a range of values within which the true population parameter is likely to lie, offering both an estimate and the uncertainty around it. They are significant because they give more information than a single point estimate, helping assess the precision and reliability of statistical conclusions, and they allow researchers to make probabilistic statements about parameters in a clear and interpretable way.

Q8. What is the concept of expected value in a probability distribution?

Ans. The expected value of a probability distribution is the long-run average or mean value you would expect if a random experiment were repeated many times.

Q9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Q10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.

Write the Python code to compute the mean sales and its confidence interval

Ans. To apply the Central Limit Theorem, we treat the daily sales data as a sample from the larger population of all sales days. We first calculate the sample mean and standard deviation, and because the sampling distribution of the mean becomes approximately normal, we can use it to estimate a 95% confidence interval.

Confidence interval = sample mean ± (critical t-value × standard deviation divided by the square root of sample size).

```python
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)
n = len(data)

t_critical = stats.t.ppf(0.975, df=n-1)

margin_of_error = t_critical * (std_sales / np.sqrt(n))

lower = mean_sales - margin_of_error
upper = mean_sales + margin_of_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower, upper))
```
