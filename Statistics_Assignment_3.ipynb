Q1. What is hypothesis testing in statistics?

Ans1. Hypothesis testing is a statistical method used to decide whether there is enough evidence in a sample to support a claim about a population. It compares a null hypothesis (no effect or no difference) with an alternative hypothesis (there is an effect or difference) using sample data, and the decision is usually based on the p-value.

Q2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

Ans2. The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population it represents the default or baseline assumption. The alternative hypothesis is the opposite; it claims that there is a real effect or a significant difference.

Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans3. The significance level (α) is a cutoff value used to decide whether the results of a hypothesis test are strong enough to reject the null hypothesis. It is usually set at 0.05 or 0.01. If the p-value from the test is smaller than α, we conclude that the evidence is strong and reject the null hypothesis. If the p-value is larger than α, we keep the null hypothesis because the evidence is not strong enough.

Q4. What are Type I and Type II errors? Give examples of each.

Ans4. Type I and Type II errors are two kinds of mistakes that can occur in hypothesis testing. A Type I error happens when we conclude that there is an effect or difference when in reality there is none for example, deciding a new medicine works better even though it actually doesn’t. A Type II error occurs when we fail to detect a real effect for instance, concluding that a new medicine does not work better when it actually does. Together, these errors highlight the risks involved in making decisions based on sample data.

Q5. What is the difference between a Z-test and a T-test? Explain when to use
each.

Ans5. A Z-test and a T-test are both used to compare sample data to a population or to compare two groups, but they differ mainly in the information available and the sample size. A Z-test is used when the population standard deviation is known and the sample size is large (typically n > 30). It assumes the sampling distribution is approximately normal. In contrast, a T-test is used when the population standard deviation is unknown and must be estimated from the sample, which is common in real situations, especially with small sample sizes (n ≤ 30).

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

plt.hist(sample_data, bins=11, edgecolor='black')
plt.xlabel('Sample Data')
plt.ylabel('Distribution')

plt.show()

#Null hypothesis
mu_0 = 50.25

z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

alpha = 0.05
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
print("Decision:", decision)
```

Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(10)
data = np.random.normal(loc=50, scale=10, size=50)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

z = norm.ppf(0.975)

lower = sample_mean - z * (sample_std / np.sqrt(n))
upper = sample_mean + z * (sample_std / np.sqrt(n))

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (lower, upper))

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(upper, color='green', linestyle='--', label='95% CI Upper')

plt.title("Simulated Normal Distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

np.random.seed(10)
data = np.random.normal(loc=50, scale=5, size=500)

zscores = zscore(data)

plt.figure(figsize=(10, 6))
plt.hist(zscores, bins=30, edgecolor='black', alpha=0.7)

for z in [-3, -2, -1, 0, 1, 2, 3]:
    plt.axvline(z, color='red' if z == 0 else 'green', linestyle='--')
    plt.text(z, plt.ylim()[1]*0.95, f"Z={z}", ha='center', color='blue')

plt.title("Histogram of Z-scores with Standard Deviation Markers")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.show()
```
