Q10. Write a code to demonstrate type casting with list elements.

```python
# Original list with string numbers
str_numbers = ["10", "20", "30.6", "40.7", "50"]

print("Original list (strings):", str_numbers)

# Convert all elements to integers (where possible)
int_numbers = [int(float(num)) for num in str_numbers]
print("List after converting to integers:", int_numbers)

# Convert all elements to floats
float_numbers = [float(num) for num in str_numbers]
print("List after converting to floats:", float_numbers)

# Convert all elements to booleans
bool_values = [bool(num) for num in str_numbers]
print("List after converting to booleans:", bool_values)
```

Q14. Write a program to reverse a string using a while loop

```python
#Enter a string input to reverse.
first_str = input("Enter the string to reverse: ")
print(f"Entered string is {first_str}")
#Generate empty list and count the length of string given as input
rev_str = ""
index = len(first_str) - 1

while index >= 0:
    rev_str += first_str[index]
#This index -=1 will move the position to read in right to left direction
    index -= 1

print("Reversed string of input is:", rev_str)
```
