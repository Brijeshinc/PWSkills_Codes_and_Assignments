## **Theoretical Questions**


Q1. What is a parameter?

Ans. A parameter is a value that defines or controls how a system, model, or function behaves.

Q2. What is correlation? What does negative correlation mean?

Ans. Correlation is a statistical measure that shows how strongly two variables are related and how they move with respect to each other.
A negative correlation means that when one variable increases, the other decreases (they move in opposite directions).

Q3. Define Machine Learning. What are the main components in Machine Learning?

Ans. Machine Learning is a field of artificial intelligence where computers learn patterns from data and improve their performance on a task without being explicitly programmed.

The main components of Machine Learning are: data, model, features, learning algorithm, and evaluation/performance metric.

Q4. How does loss value help in determining whether the model is good or not?

Ans. The loss value measures how far the model’s predictions are from the actual values. A lower loss indicates better model performance, while a high loss means poor predictions. By tracking loss during training and validation, we can judge how well the model is learning and generalizing.

Q5. What are continuous and categorical variables?

Ans. Continuous variables are numerical values that can take any value within a range (e.g., height, weight, temperature). Categorical variables represent distinct groups or labels with no inherent numerical meaning (e.g., gender, color, blood type).

Q6. How do we handle categorical variables in Machine Learning? What are the common techniques?

Ans. Categorical variables are handled by converting them into numerical form so ML models can process them. Common techniques include label encoding, one-hot encoding, ordinal encoding, and target/frequency encoding.

Q7. What do you mean by training and testing a dataset?

Ans. Training a dataset means using data to teach the model by learning patterns and parameters. Testing a dataset means evaluating the trained model on unseen data to measure its performance and generalization.

Q8. What is sklearn.preprocessing?

Ans. sklearn.preprocessing is a module in Scikit-learn used to prepare and transform data before applying machine learning models. It provides tools for scaling, normalizing, and encoding features so models perform better.

Q9. What is a Test set?

Ans. A test set is a portion of the dataset kept aside and not used during training. It is used to evaluate the final performance of a model on unseen data and check how well it generalizes.

Q11. Why do we have to perform EDA before fitting a model to the data?

Ans. EDA is performed to understand the data before modeling. It helps identify patterns, relationships, missing values, outliers, and errors, ensuring better feature selection and improved model performance.

Q12. What is correlation?

Ans. Correlation is a statistical measure that shows the strength and direction of the relationship between two variables.

Q13. What does negative correlation mean?

Ans. Negative correlation means that two variables move in opposite directions. As one variable increases, the other decreases.

Q14. How can you find correlation between variables in Python?

Ans. Correlation between variables in Python can be found using .corr() in pandas or corrcoef() from NumPy. It returns a correlation value that indicates the strength and direction of the relationship.

Q15. What is causation? Explain difference between correlation and causation with an example.

Ans. Correlation means that two variables are related and change together, but it does not imply cause and effect. Causation means that a change in one variable directly produces a change in another.
Correlation only shows association, while causation explains the reason behind the relationship.

Q16. An optimizer is an algorithm used to update model parameters (weights and biases) in order to minimize the loss function during training.

Types of optimizers:

Gradient Descent: Updates parameters using the full dataset; e.g., linear regression training.

Stochastic Gradient Descent (SGD): Updates parameters using one data point at a time; faster and widely used.

Mini-batch Gradient Descent: Uses small batches of data; balances speed and stability.

Momentum: Accelerates SGD by considering past gradients; used in deep neural networks.

Adam (Adaptive Moment Estimation): Combines momentum and adaptive learning rates; commonly used in deep learning models.

Q17. What is sklearn.linear_model?

Ans. sklearn.linear_model is a module in Scikit-learn that provides linear models for regression and classification tasks. It includes algorithms like Linear Regression, Logistic Regression, Ridge, Lasso, and ElasticNet used to model relationships between variables.

Q18. What does model.fit() do? What arguments must be given?

Ans. model.fit() trains the machine learning model by learning patterns and parameters from the data. It requires input features (X) and target/output values (y) as arguments.

Q19. What does model.predict() do? What arguments must be given?

Ans. model.predict() uses the trained model to make predictions on new or unseen data. It requires the input features (X) as its argument and returns the predicted values.

Q20. What are continuous and categorical variables?

Ans. Continuous variables are numerical values that can take any value within a range (e.g., height, weight, temperature). Categorical variables represent distinct groups or labels with no numerical meaning (e.g., gender, color, blood group).

Q21. What is feature scaling? How does it help in Machine Learning?

Ans. Feature scaling is the process of bringing all features to a similar scale.
It helps machine learning models converge faster and perform better, especially distance-based and gradient-based algorithms.

Q22. How do we perform scaling in Python?

Ans. Scaling in Python is done using preprocessing tools from Scikit-learn.
Common methods include StandardScaler, MinMaxScaler, and RobustScaler from sklearn.preprocessing.

Q23. What is sklearn.preprocessing?

Ans. sklearn.preprocessing is a module in Scikit-learn used to prepare and transform data before training machine learning models. It provides tools for feature scaling, normalization, and encoding categorical variables.

Q24. How do we split data for model fitting (training and testing) in Python?

Ans. Data is split using train_test_split() from sklearn.model_selection.
It divides the dataset into training and testing sets to evaluate model performance on unseen data.

Q25. Explain data encoding?

Ans. Data encoding is the process of converting categorical data into numerical form so machine learning models can understand it. Common methods include label encoding, one-hot encoding, and ordinal encoding.
