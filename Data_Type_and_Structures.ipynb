Q2.Explain the difference between mutable and immutable data types with examples.

Ans2. Mutable data type can be changed or modified after assigning memory (example - list, dict, set), whereas immutable data types can not be changed or modified once assigned memory (example - tuple, string).

Q8. What is a hash table, and how is it relate to dictionaries in python?

Ans8. A hash table is a data structure that stores data in key value pairs and uses a hash function to compute an index for each key. Each dictionary key is passed through a hash function to generate an index, which determines where the corresponding value is stored in memory.

Q10. Explain why string are immutable in Python.

Ans10. Strings are immutable in Python because once a string is created, its value cannot be changed. This immutability allows Python to optimize memory usage and ensures that strings can be safely used as keys in dictionaries or elements in sets without risk of accidental modification.

Q11. What advantages do dictionaries offer over lists for certain tasks?

Ans11. Dictionaries offer faster lookups because they use a hash table to access values by keys, unlike lists which require scanning elements sequentially. They also allow direct access via unique keys, making them ideal for tasks like counting items, mapping relationships, which would be slower or more complex with lists.

Q17. Describe the time complexity of accessing elements in a dictionary.

Ans17. Time complexity in a dictionary refers to how long it takes to perform operations like accessing, inserting, or deleting a key value pair relative to the size of the dictionary.
For operations, dictionaries have average time complexity, meaning they take constant time regardless of the number of items.

Q18. In what situations are lists preferred over dictionaries?

Ans18. Lists are useful when we need to store a sequence of items and access them by index rather than by key. They are also simpler and use less memory than dictionaries when key value mapping is not required.

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Ans19. Dictionaries in Python are considered unordered because they do not store elements in the order we add them. If we loop over a dictionary, the keys may appear in a different order than how we inserted them.
