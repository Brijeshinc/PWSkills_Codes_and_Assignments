## Theoretical Questions

Q1. What is Object Oriented Programming (OOP)?

Ans. Object-Oriented Programming (OOP) in Python is a way of programming that models real-world things as objects, which bundle together both data (like characteristics) and actions (like behaviors). It helps make programs structured, reusable, and easier to understand.

Q2. What is a class in OOPs?

Ans. In OOP, a class is a blueprint or template for creating objects.

Q3. What is an Object in OOPs?

Ans. In OOP, an object is a real-world entity that contains data (attributes) and behavior (methods).

Q4. What is difference between abstraction and encapsulation?

Ans. Encapsulation is about hiding the internal state of an object and restricting direct access.
Abstraction is about hiding implementation details and showing only the essential features

Q5. What are dunder methods in python?

Ans. Dunder methods in Python are special built-in methods that start and end with __ (double underscore).

Q6. Explain the concept of inheritance in OOPs.

Ans. Inheritance in OOP is the concept where one class (derived class) can reuse and extend the properties and behaviors of another class (base class). It promotes code reusability and supports creating hierarchical relationships.

Q7. What is polymorphism in OOPs?

Ans. Polymorphism in OOP means the same function or method name can have different behaviors depending on the object that calls it.

Q8. How is encapsulation is achived in python?

Ans. Encapsulation in Python is achieved by hiding an object’s internal data using access specifiers: public, protected (_var), and private (__var).

Q9. What is Constructor in python?

Ans. A constructor in Python is a special method __init__() used to initialize an object’s attributes when it is created. It runs automatically when an instance of a class is made.

Q10. What are class and static method in python?

Ans. a class method (using @classmethod) takes the class (cls) as the first parameter and can access or modify class-level attributes. A static method (using @staticmethod) doesn’t take self or cls and cannot access instance or class data; it behaves like a regular function inside the class.

Q11. What is method overloading in python?

Ans. Method overloading is the ability to define multiple methods with the same name but different parameters in a class.

Q12. What is method overriding in OOPs?

Ans. Method overriding in OOP occurs when a child class provides its own implementation of a method that already exists in the parent class. It allows the derived class to modify or extend the behavior of the base method.

Q13. What is property of decorator in python?

Ans. Property decorator allowos us to use methods as an attributes

Q14. Why is polymorphism important in OPP?

Ans. Polymorphism is important because it allows the same method or operation to work differently on different objects, making code more flexible, reusable, and easier to maintain.

Q15. What is an abstract class in python?

Ans. An abstract class in Python is a class that cannot be instantiated and serves as a blueprint for other classes. It can contain abstract methods (declared but not implemented) that must be overridden by subclasses, ensuring a consistent interface.

Q16. What are the advantages of OPP?

Ans. Advantages of OOP include modularity, as code is organized into classes and objects, and reusability, since classes and methods can be reused. Encapsulation protects data, inheritance reduces code duplication, and polymorphism allows flexible use of the same interface for different objects, making programs easier to maintain and scale.

Q17. What is the difference between a class variable and an instance variable?

Ans. A class variable is shared among all instances of a class and is defined directly in the class body. An instance variable belongs to a specific object, created inside methods (usually __init__) using self.

Q18. What is multiple inheritance in Python?

Ans. Multiple inheritance in Python is when a class inherits from more than one parent class, allowing it to access attributes and methods of all the parents.

Q19. Explain the purpose of "__str__" and "__repr__" methods in Python.

Ans. The __str__ method defines a human-readable string representation of an object, mainly used with print() or str(). The __repr__ method provides an unambiguous string representation, mainly for developers, and is used in the interpreter or with repr()

Q20. What is the signaficance of super() function in python?

Ans. The super() function in Python is used to call methods from a parent (super) class inside a child class. It helps in reusing code, avoiding duplication, and supporting multiple inheritance

Q21. What is the significance of the __del__ method in Python?

Ans. The __del__ method in Python is a destructor that is called when an object is about to be destroyed. It is typically used to release resources like closing files or database connections.

Q22. What is the difference between @staticmethod and @classmethod in Python?

Ans. @staticmethod is a method that does not take self or cls as the first argument and cannot access instance or class data — it behaves like a normal function inside the class. Whereas @classmethod takes cls as the first argument and can access or modify class variables, making it useful for operations that affect the class as a whole rather than individual objects.

Q23. How does polymorphism work in Python with inheritance?

Ans. Polymorphism with inheritance means a child class can provide its own implementation of a method defined in the parent class. When the method is called on an object, Python automatically decides which version to run based on the object’s class.

Q24.  What is method chaining in Python OOP?

Ans. Method chaining in Python OOP is a technique where multiple methods are called on the same object in a single line, one after another. This is achieved by having each method return self.

Q25. What is the purpose of the __call__ method in Python?

Ans. The __call__ method in Python allows an object of a class to be called like a function. When defined, you can use parentheses () after an object, and Python will execute the code inside __call__.

Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Drive a class Car and further drive a class Electric Car that adds a battery attribute.

```python
class Vehicle:

    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def show_type(self):
        print(f"Vehicle type: {self.type}")


class Car(Vehicle):

    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def show_brand(self):
        print(f"Car brand: {self.brand}")


class ElectricCar(Car):

    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery

    def show_battery(self):
        print(f"Battery capacity: {self.battery} kWh")



ecar = ElectricCar("Car", "Tesla", 100)
ecar.show_type()
ecar.show_brand()
ecar.show_battery()
```

Q5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance, and method to deposit, withdraw and check balance.

```python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")
    def check_balance(self):
        return self.__balance

act = BankAccount(5000)
act.deposit(500)
act.withdraw(200)
print(act.check_balance())
```

```python
class Vector:
    def __init__(self, x, y):

        self.x = x
        self.y = y

    def __add__(self, other):

        return Vector(self.x + other.x, self.y + other.y)



v1 = Vector(1, 2)
v2 = Vector(3, 4)

v3 = v1 + v2
print(v3.x, v3.y)
```

Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus



emp = Employee("Arnav", 40, 20)
mgr = Manager("Ajit", 40, 30, 500)

print(f"{emp.name}'s Salary: {emp.calculate_salary()}")   # 800
print(f"{mgr.name}'s Salary: {mgr.calculate_salary()}")   # 1700
```

Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

```python
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price
    def details(self):
        return f"Address: {self.address}\nPrice: {self.price}"

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        self.number_of_rooms = number_of_rooms
        super().__init__(address, price)
    def details(self):
        return f"Address: {self.address}\nPrice: {self.price}\nNumber of Rooms: {self.number_of_rooms}"

H1 = House("5, Privet Drive, Little Whinging, Surrey, England,", 5000000)
print(H1.details())

H2 = Mansion("5, Privet Drive, Little Whinging, Surrey, England,", 5000000, 5)
print(H2.details())
```
