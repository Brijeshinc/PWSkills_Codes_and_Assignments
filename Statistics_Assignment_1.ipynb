Q1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Ans. Descriptive statistics summarize and describe the features of a dataset using measures like mean, median, or mode.

Example - A class of 40 students has an average height of 160 cm

Inferential statistics use a sample of data to make predictions or generalizations about a larger population, such as estimating averages or testing hypotheses.

Example - Measuring the height of 50 students to estimate the average height of all students in the school

Q2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

Ans. Sampling is the method of selecting a smaller group (sample) from a large population to study and draw conclusions about the entire population. In random sampling, every member of the population has an equal chance of being chosen, making it simple and unbiased for example, selecting students through a lottery system. In stratified sampling, the population is divided into specific groups or strata based on characteristics such as age, gender, or class, and samples are taken proportionally from each group to ensure better representation. This method is more precise when the population has clear subgroups.

Q3.  Define mean, median, and mode. Explain why these measures of central
tendency are important.

Ans. Mean is the average of all values (sum of values / number of values).
Median is the middle value when the data is arranged in order. Mode is the value that appears most frequently in the dataset. These measures are important because they help summarize large sets of data with a single representative value, making it easier to understand patterns, compare groups, and make decisions based on the overall trend.

Q4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

Ans. Skewness measures the asymmetry of a distribution. If data are perfectly symmetric, skewness is zero. A positive or negative skew shows that one tail of the distribution is longer than the other. Kurtosis measures the peakedness or heaviness of the tails* of a distribution. High kurtosis means heavy tails and sharp peak, while low kurtosis means light tails and a flatter peak. A positive skew implies that the data have a long tail on the right side, meaning most values are concentrated on the left, with a few unusually large values pulling the tail to the right.

Q5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = mean(numbers)
median_value = median(numbers)
mode_value = mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

Q6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

covariance = np.cov(x, y, ddof=1)[0][1]

correlation = np.corrcoef(x, y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

Q7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()
```

The boxplot shows the distribution of the dataset by highlighting its central tendency and variability. The box, which ranges from the first quartile (around 18) to the third quartile (around 24), represents the middle 50% of the data, indicating that most values lie within this range. The median line inside the box is near 21–22, showing that half of the values fall below this point and half above it. The whiskers extend from about 12 to 29, covering the spread of the non-outlier values in the dataset. A single point at 35 appears above the upper whisker, indicating an outlier meaning this value is significantly higher than the rest of the data.

Q8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.


*   Explain how you would use covariance and correlation to explore this
relationship.
*   Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]

daily_sales = [2200, 2450, 2750, 3200, 4000]






```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

x = np.array(advertising_spend)
y = np.array(daily_sales)

correlation = np.corrcoef(x, y)[0][1]

print("Correlation Coefficient:", correlation)
```

We can use covariance and correlation. Covariance tells you the direction of the relationship whether both variables increase together or move oppositely but it doesn't indicate the strength. Correlation, however, measures both the direction and strength of the relationship on a scale from –1 to 1. A value close to 1 means a strong positive relationship, meaning higher advertising spend is associated with higher sales. By calculating correlation, the marketing team can understand how strongly spending influences sales trends. This correlation value (close to 1) will show a strong positive relationship, indicating that as advertising spending increases, daily sales also rise significantly.

Q9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.

*   Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use
*   Write Python code to create a histogram using Matplotlib for the survey data:

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]



```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=5)
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Scores")
plt.ylabel("Frequency")
plt.show()
```

To understand the distribution of customer satisfaction scores before launching the new product, we would first compute summary statistics such as the mean, median, and standard deviation to get an overall sense of customer sentiment and variability. In this case, the scores cluster around 7–9, indicating generally high satisfaction with a few lower scores, helping the team gauge overall customer sentiment before launching the new product.
